```python
# list available attribution methods
inseq.list_feature_attribution_methods()
```

```python
prompt = "Hello, hope you're enjoying this explainability"
```

```python
# load a model from HuggingFace model hub and define the feature attribution 
# method you want to use
mdl = inseq.load_model("gpt2", "integrated_gradients")
```

```python
help(inseq.load_model)
```

```python
help(mdl.attribute)
```

```python
# compute the attributions for a given prompt
attr = mdl.attribute(
    prompt,
    generation_args={"max_new_tokens": 20},
    n_steps=500,
    internal_batch_size=50 )

attr.show()
```

```python
mdl = inseq.load_model("gpt2", "saliency")


attr = mdl.attribute(
    prompt,
    generation_args={"max_new_tokens": 20},
    n_steps=500,
    internal_batch_size=50 )

attr.show()
```
